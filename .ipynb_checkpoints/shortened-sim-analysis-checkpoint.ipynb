```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
!ls ../mini-sims
```

```python
lengths = ['1', '40', '79', '118', '157', '196']
```

```python
viscos = pd.DataFrame()
for ns in lengths:
    viscos[f'{ns}-visc'] = pd.Series(np.load(f'../mini-sims/{ns}ns_sim.npz', allow_pickle=True)['arr_3'])
    viscos[f'{ns}-auto'] = pd.Series(np.load(f'../mini-sims/{ns}ns_sim.npz', allow_pickle=True)['arr_2'])
    viscos[f'{ns}-time'] = pd.Series(np.load(f'../mini-sims/{ns}ns_sim.npz', allow_pickle=True)['arr_1'])
```

```python
viscos.head()
```

```python
plt.figure(figsize=(12,7))
for c, ns in enumerate(lengths):
    plt.plot(viscos[f'{ns}-time']/1000, viscos[f'{ns}-visc'], c = plt.cm.viridis(plt.Normalize(0,6)(c)))
plt.ylim(0,1e-10)
plt.xlim(0,50)
```

```python
plt.figure(figsize=(12,7))
for c, ns in enumerate(lengths):
    plt.plot(viscos[f'{ns}-time'], viscos[f'{ns}-auto'], c = plt.cm.viridis(plt.Normalize(0,6)(c)), linewidth=c)
plt.ylim(-500,2000)
# plt.yscale('symlog')
# plt.yscale('log')
plt.xlim(np.array([-0.1,1])*10)
```
