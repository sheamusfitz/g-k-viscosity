```python
data = pd.DataFrame()

lipids = sp.array(['dopc-293','dopc-303','dopc-313','dopc-323','dmpc','dppc','sopc','popc'])

# basename = '/py_output_noshift.npz'
basename = '/py_output_shift.npz'

for l, lip in enumerate(lipids):
    for r in range(5):
#         print(lip,r+1)
        filename = '../'+lip+'/r'+str(r+1)+basename
        temp = sp.load(filename)
        data[lip+'-r'+str(r+1)+'-visc'] = sp.array(temp.f.arr_1)
        data[lip+'-r'+str(r+1)+'-visc-unc'] = sp.absolute(sp.array(temp.f.arr_2))
        data[lip+'-r'+str(r+1)+'-visc-unc'][0] = data[lip+'-r'+str(r+1)+'-visc-unc'][1]
        data[lip+'-r'+str(r+1)+'-roll'] = sp.array(temp.f.arr_3)
        data[lip+'-r'+str(r+1)+'-roll-unc'] = sp.array(temp.f.arr_4)
        data[lip+'-r'+str(r+1)+'-roll-unc'][0:2] = data[lip+'-r'+str(r+1)+'-roll-unc'][2]
        data[lip+'-r'+str(r+1)+'-autocor'] = sp.array(temp.f.arr_5[:,0])
        data[lip+'-r'+str(r+1)+'-autocor-unc'] = sp.array(temp.f.arr_5[:,1])
x = sp.arange(len(data['dppc-r1-roll']))*2e-12
data
```

```python
# datatofit = data['dopc-293-r4-autocor']/data['dopc-293-r4-autocor'][0]
# errtofit = data['dopc-293-r4-autocor-unc']/data['dopc-293-r4-autocor'][0]

d1 = data['dopc-293-r1-autocor']
d2 = data['dopc-293-r2-autocor']
d3 = data['dopc-293-r3-autocor']
d4 = data['dopc-293-r4-autocor']
d5 = data['dopc-293-r5-autocor']

e1 = sp.absolute(data['dopc-293-r1-autocor-unc'])
e2 = sp.absolute(data['dopc-293-r2-autocor-unc'])
e3 = sp.absolute(data['dopc-293-r3-autocor-unc'])
e4 = sp.absolute(data['dopc-293-r4-autocor-unc'])
e5 = sp.absolute(data['dopc-293-r5-autocor-unc'])

datatofit = sp.average(sp.array([d1,d2,d3,d4,d5]),axis=0, weights = sp.array([e1,e2,e3,e4,e5])**-2)
# print(datatofit)

errtofit = (
    (
    sp.average(sp.power([d1,d2,d3,d4,d5],2), axis=0, weights = sp.array([e1,e2,e3,e4,e5])**-2)
    - sp.average(sp.array([d1,d2,d3,d4,d5]),axis=0, weights = sp.array([e1,e2,e3,e4,e5])**-2)**2
    )/4
)**0.5

scaling = datatofit[0]

errtofit /= scaling
datatofit /= scaling

errtofit = sp.real_if_close(errtofit)


plt.plot(x, datatofit,'.')
# plt.plot(x,d1,'-')
plt.ylim(-0.01,0.01)
plt.xlim(-1e-11,5e-10)
plt.show()
plt.plot(x, (errtofit/datatofit),'-')
plt.yscale('symlog')
plt.show()
plt.plot(errtofit,',')
plt.plot(e1/scaling)
plt.plot(e2/scaling)
plt.plot(e3/scaling)
plt.plot(e4/scaling)
plt.plot(e5/scaling)
plt.yscale('log')
```

Once again, that's
$$s_{wtd} = \sqrt{\left(    \frac{\sum_{i=1}^n w_ix_i^2}{\sum_{i=1}^n w_i} - \left(    \bar x_{wtd}    \right)^2   \right)\frac{1}{n-1}}$$

```python
averaged_viscs = pd.DataFrame()

for lipid in lipids:
    viscs = sp.array([data[lipid+'-r1-visc'], data[lipid+'-r2-visc'], data[lipid+'-r3-visc'],
                      data[lipid+'-r4-visc'], data[lipid+'-r5-visc']])
    weights = sp.array([data[lipid+'-r1-visc-unc'], data[lipid+'-r2-visc-unc'], data[lipid+'-r3-visc-unc'],
                      data[lipid+'-r4-visc-unc'], data[lipid+'-r5-visc-unc']])**(-2)
    averaged_viscs[lipid+'-visc'] = sp.average(viscs, weights=weights, axis=0)
    averaged_viscs[lipid+'-unc'] = (
        (
        sp.sum(weights*viscs**2, axis=0)/sp.sum(weights, axis=0)-averaged_viscs[lipid+'-visc']**2
        )/4
    )**(1/2)
averaged_viscs
```

```python
integrated_new = sp.array(
    [sp.sum(newy[: i + 1]) for i in range(len(newy))]
)
integrated = sp.array(
    [sp.sum(oldy[: i + 1]) for i in range(len(newy))]
)

unclist_new = sp.array(
    [sp.sqrt(sp.sum(err[:i] ** 2))
        for i in range(1, len(newy))]
)
unclist_new = sp.insert(unclist_new, 0, unclist_new[0])
unclist_new = sp.array(unclist_new, dtype=float)
print('a')
unclist = sp.array(
    [sp.sqrt(sp.sum(oerr[:i] ** 2))
        for i in range(1, len(oldy))]
)
unclist = sp.insert(unclist, 0, unclist[0])
unclist = sp.array(unclist, dtype=float)
```
