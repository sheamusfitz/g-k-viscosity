```python
import pandas as pd
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt
```

```python
test_in = sp.load('../dopc-293/r1/py_output.npz')
test_in.f.arr_0
```

```python
data = pd.DataFrame()

lipids = ['dopc-293','dopc-303','dopc-313','dopc-323','dmpc','dppc','sopc','popc']

for l, lip in enumerate(lipids):
    for r in range(5):
        if lip!='dppc' and r!=2:
            filename = '../'+lip+'/r'+str(r+1)+'/py_output.npz'
            temp = sp.load(filename)
            data[lip+'-r'+str(r+1)+'-visc'] = temp.f.arr_1
            data[lip+'-r'+str(r+1)+'-visc-unc'] = temp.f.arr_2
            data[lip+'-r'+str(r+1)+'-roll'] = temp.f.arr_3
            data[lip+'-r'+str(r+1)+'-roll-unc'] = temp.f.arr_4
data
```

```python
for l, lipid in enumerate(lipids):
    scolor = 'C'+str(l)
    for r in range(5):
        if lipid!='dppc' and r!=2:
            plt.plot(data[lipid+'-r'+str(r+1)+'-visc'], c=scolor)
    plt.show()
```
